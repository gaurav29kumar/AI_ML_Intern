```python
import pickle
import pandas as pd
import numpy as np

with open('housing_linear_model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)

# Taking Input
data_input = pd.DataFrame([{
    'MedInc': 6.0, 'HouseAge': 10.0, 'AveRooms': 6.2, 'AveBedrms': 1.0,
    'Population': 850.0, 'AveOccup': 2.5, 'Latitude': 37.77, 'Longitude': -122.41
}])

# 3. Predict using the loaded model
prediction = loaded_model.predict(data_input)
# Flattening the array in 1d to get single float value
single_num=float(np.atleast_1d(prediction))
print(f"Predicted Price from Loaded Model: ${single_num * 100000:,.2f}")
```
